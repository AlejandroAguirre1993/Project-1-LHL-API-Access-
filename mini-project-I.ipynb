## Task
What are the different modes of transport which are operated by Transfer for London? How many of modes do they have?

Print the list with different modes of transport, plus their count. Example output:
```
[bus, cable-car,.....]
Number of different modes of transport is: xyz
```

We need to search the documentation for correct request.

````python
urltransport = "https://api.tfl.gov.uk/Line/Meta/Modes"
````

````python
transport_request = requests.get(urltransport, headers={'accept': 'application/JSON'}).json()
##### 'modeName' is the value to access to know wwhat type of vehicle
List_of_modes = []
for transport in transport_request:
    Mode = transport['modeName']
    List_of_modes.append(Mode)
number_of_modes = len(List_of_modes)   

print(List_of_modes)
print("Number of different modes of transport is:", + number_of_modes)
````

## Task
How many BikePoints in London are operated by Transfor for London? How many docks are in **all** BikePoints? There is the information for empty and full docks for each BikePoint.

````python
urlbikes = "https://api.tfl.gov.uk/BikePoint"
````

````python
###Send request to API
Bike_request = requests.get(urlbikes, headers={'accept': 'application/JSON'}).json()
#### Initialize variables to count number of Bikes and Docks
Num_Bikes = 0
Num_Docks = 0

for Bkps in Bike_request:
    bikes = int(Bkps["additionalProperties"] [6] ["value"]) # the value is giving as a string, look at documentation and visualize the Bike_request to find indexing
    docks = int(Bkps["additionalProperties"] [8] ["value"]) # indexing 8 is for docks
    Num_Bikes = bikes + Num_Bikes
    Num_Docks = docks + Num_Docks

print("Number of bikes at London:", +  Num_Bikes)  
print("Number of docks at London:", +  Num_Docks)  
    
  
````

## Task
How many tube and bus lines are in London? Print names of all tube lines.

````python
url_tube_bus = "https://api.tfl.gov.uk"
mode = "tube, bus"
plus_url = f"/Line/Mode/{mode}"
````

````python
###Send request to API
Lines_request = requests.get(url_tube_bus + plus_url, headers={'accept': 'application/JSON'}).json()
#### Initialize variables to count number of tube lines and bus
Num_tubes = 0
Num_buses = 0
List_names_tblines = []
for lines in Lines_request:
    if lines['modeName'] == 'bus':
        Num_buses += 1
    else:
        Num_tubes += 1
        name_tb_line = lines['name']
        List_names_tblines.append(name_tb_line)
print("Number of buses lines:", Num_buses)
print("Number of tubes lines:", Num_tubes)
print(List_names_tblines)
````

## Task
How many station has `victoria` line?

````python
url_default = "https://api.tfl.gov.uk"
id = "victoria"
victoria_url = f"/Line/{id}/StopPoints"

````

````python
###Send request to API
Victoria_request = requests.get(url_default + victoria_url, headers={'accept': 'application/JSON'}).json()
define_victoria = 'victoria'
##Initialize Victoria counter###
Victoria_lines_counter = 0
for vline  in Victoria_request:
    if  define_victoria in str(vline['lineModeGroups']) :
        Victoria_lines_counter += 1
        print(vline['commonName'])
print("Number of station that has Victoria Line: ", + Victoria_lines_counter )
````
